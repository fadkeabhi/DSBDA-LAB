```python
"""
Problem Statement No. 11 
1. Use the dataset 'titanic'. Plot a box plot for distribution of age with respect to each gender along with the information 
about whether they survived or not. (Column names : 'sex' and 'age')  
2. Write observations on the inference from the above statistics. Use violin plot and others plots from seaborn package                                               
Use :Titanic.csv  
"""
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop('PassengerId',axis=1,inplace=True)
```

```python
df.drop('Name',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.info()
```

```python
import seaborn as sns
```

```python
sns.boxplot(x='Age',y='Sex',hue='Survived',data=df)
```

```python
sns.violinplot(x='Age',y='Sex',hue='Survived',data=df,split=True)
```

```python
sns.swarmplot(x='Age',y='Sex',hue='Survived',data=df)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sex', y='Age', hue='Survived', data=df)
plt.title('Age Distribution by Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Sex', y='Age', hue='Survived', data=df)
plt.title('Age Distribution by Gender and Survival (Swarm Plot)')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Sex', y='Age', hue='Survived', data=df)
plt.title('Age Distribution by Gender and Survival (Swarm Plot)')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```
