```python
import matplotlib.pyplot as plt
```

<h1>Histogram</h1>

```python
plt.hist(df['SepalLengthCm'],bins=10,color='blue',edgecolor='black')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.title('Histogram of SepalLength in Centimeter')
plt.show()
```

```python
plt.hist(df['PetalLengthCm'],bins=10,color='pink',edgecolor='red')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length in Centimeter')
```

```python
plt.hist(df['SepalWidthCm'],bins=10,color='blue',edgecolor='black')
```

```python
plt.hist(df['PetalWidthCm'],bins=10,color='black',edgecolor='white')
```

```python
fig , axis = plt.subplots(nrows = 2 ,ncols = 2 , figsize = (12,8))
df['PetalWidthCm'].plot(kind='hist',bins=10,color='pink',edgecolor='red',ax = axis[0,0])
axis[0,0].set_xlabel('Length in CM')
axis[0,0].set_ylabel('Frequency')
axis[0,0].set_title('Petal Width in CM')

df['SepalWidthCm'].plot(kind='hist',bins=10,color='black',edgecolor='white',ax= axis[0,1])
axis[0,1].set_xlabel('Length in CM')
axis[0,1].set_ylabel('Frequency')
axis[0,1].set_title('Sepal Width in CM')

df['SepalLengthCm'].plot(kind = 'hist',bins=10,color='blue',edgecolor='white',ax=axis[1,0])
axis[1,0].set_xlabel('Length in CM')
axis[1,0].set_ylabel('Frequency')
axis[1,0].set_title('Sepal Length in CM')

df['PetalLengthCm'].plot(kind = 'hist',bins=10,color='red',edgecolor='white',ax=axis[1,1])
axis[1,1].set_xlabel('Length in CM')
axis[1,1].set_ylabel('Frequency')
axis[1,1].set_title('Petal Length in CM')

plt.tight_layout()
plt.show()
```

<h1>Box Plot</h1>

```python
plt.boxplot(df['PetalWidthCm'])
plt.xlabel('')
plt.ylabel('Length in CM')
plt.title('Boxplot of Petal width')
plt.show()
```
