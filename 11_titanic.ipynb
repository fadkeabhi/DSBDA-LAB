## Problem Statement No. 11

1. Use the dataset 'titanic'. Plot a box plot for distribution of age with respect to each gender along with the information about whether they survived or not. (Column names : 'sex' and 'age') 
2. Write observations on the inference from the above statistics. Use violin plot and others plots from seaborn package

Use :Titanic.csv

```python
import pandas as pd
```

```python
df = pd.read_csv("Datasets/titanic.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop(["Cabin"], axis=1, inplace=True)
```

```python
age_avg = int(df["Age"].mean())
age_avg
```

```python
df["Age"] = df["Age"].apply(lambda x: age_avg if pd.isnull(x) else x)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True,ignore_index=True)
```

```python
df
```

```python
import seaborn as sns
```

```python
sns.boxplot(data=df, x="Age", y="Sex")
```

```python
sns.boxplot(data=df, x="Age", y="Sex", hue="Survived")
```

- The box plot shows the distribution of ages for male and female passengers, grouped by survival status.
- For both males and females, the median age of passengers who survived appears to be lower than the median age of passengers who did not survive.
- Younger passengers were more likely to survive. 
- There are more outliers in the age distribution of non-survivors, particularly for males, suggesting that there were more older passengers among the non-survivors.

```python
sns.violinplot(data=df, x="Sex", y="Age", hue="Survived")
```

```python
sns.swarmplot(data=df, x="Sex", y="Age", hue="Survived")
```

```python
sns.swarmplot(data=df, x="Sex", y="Age", hue="Survived",size=1)
```
